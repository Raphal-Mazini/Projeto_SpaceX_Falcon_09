```python
df_falcon = pd.read_csv("/content/DbApi_bruto_falcon_09.csv")
df_falcon.head()
```

### Verificação de valores nulo/ausentes

```python
df_falcon.isnull().sum()
```

Pode se observar em duas variaveis, uma com 5 campos com ausencia de valor e outra, com 26 campos com ausencia de valor.

Será iniciado com a variavel, ``PayloadMass``, calcula se a média da coluna e substituísse os valores nulos pela média.

```python
# Calcular a média da variavel "PayloadMass"
mean = df_falcon['PayloadMass'].mean()

# Substitir os valores np.nan pela média
df_falcon['PayloadMass'] = df_falcon['PayloadMass'].fillna(mean)
```

Agora não deve ter valores ausentes no conjunto de dados, exceto em ``LandingPad``.

```python
# Após tratamento dos valores nulos
df_falcon.isnull().sum()
```
